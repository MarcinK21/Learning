### Wyszukiwanie liczby oraz jej indeksu w liście z losowo wygenerowanymi liczbami

```python
import random

# Tworzenie pustej listy do przechowywania losowych liczb
lista = []

# Wypełnienie listy 100 losowymi liczbami całkowitymi z zakresu od 100 do 999
for i in range(100):
  lista.append(random.randint(100,999))

#print(lista)

# Pobranie liczby od użytkownika i konwersja na typ całkowity
n = int(input("Podaj liczbe:"))

# Gdyby liczba podana przez użytkownika nie została znaleziona, zostanie wyświetlony komunikat:
a = "Nie ma takiej liczby"

# Przeszukiwanie listy w celu znalezienia liczby podanej przez użytkownika
for j in lista:
  if j == n:
    # Jeśli liczba zostanie znaleziona, przygotowanie wiadomości z jej indeksem
    b = str(lista.index(n))
    a = "Liczba znajduje się na liście pod indeksem " + b + "."

# Wyświetlenie wyniku wyszukiwania
print(a)
```

### Tworzenie listy rosnącego ciągu liczb

```python
from random import randrange

def increasing_list(how_many):
    tab = [0]
    
    for i in range(1, how_many):
        tab.append(tab[i - 1] + randrange(5))
    return tab
```

```python
Ilist = increasing_list(100)

# print(Ilist)     # w razie jakbym chciał sprawdzić jak wygląda stworzona lista
```

### Wyszukiwanie binarne w posortowanej liście

```python
def BinarySearch(tab, szukana):

  od = 0
  do = len(tab)
  indeks = 0 

# pętla dopóki zakres nie zostanie zawężony do 0
  while od <= do :
    indeks = (od + do) // 2   # Środkowy indeks zakresu
    
    # Porównanie środkowego elementu z szukanym
    if tab[indeks] > szukana : 
      do = indeks - 1  # Zawężenie zakresu wyszukiwania do "dolnej połowy"
      
    elif tab[indeks] < szukana :
      od = indeks + 1 # Zawężenie zakresu wyszukiwania do "górnej połowy"
      
    else :
      return indeks  
    
  return None
```

```python
BinarySearch([2,4,6,8,10,12,14,16], 14)
```

### Wyszukiwanie binarne w posortowanej liście - rekurencja

```python
def BinarySearchR(tab, szukana, od=0, do=-3):
  # Ustalenie domyślnej wartości, jeśli nie zostanie podana
  if do == -3 :
    do = len(tab)-1 

# Sprawdzenie czy zakres jest prawidłowy
  if od > do :
    return -1

  mid = (od + do) // 2  # Oblicza środkowy indeks zakresu
  if tab[mid] == szukana: # Sprawdzenie czy środkowy element jest szukaną wartością
    return mid

  if tab[mid] < szukana: # Decyzja, którą połowę listy należy dalej przeszukiwać
    return BinarySearchR(tab, szukana, mid+1, do)
  
  return BinarySearchR(tab, szukana, mid-1, do)
```

```python
BinarySearchR([2,4,6,8,10,12,14,16], 14, od=0, do=-3)
```
