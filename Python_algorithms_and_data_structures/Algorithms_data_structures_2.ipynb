### Selection Sort
- sortowanie przez wybieranie

```python
def SelectionSort(tab):
    # Przechodzi przez każdy element z listy "tab"
    for i in range(len(tab)): 
        # Przeszukiwanie pozostałej części listy w celu znalezienia najmniejszego elementu
        for j in range(i+1, len(tab)):
            # Jeśli znaleziono element mniejszy niż obecnie rozważany
            if tab[j] < tab[i]:
                # Zamiana elementów miejscami
                tab[i], tab[j] = tab[j], tab[i]
                
    return tab
```

```python
SelectionSort([7,15,2,8,7,3,1,9])
```

### Selection Sort
- "ulepszona wersja"

```python
def SelectionSortB(tab):
    for i in range(len(tab)-1):
        minimum = i
        for j in range(i+1,len(tab)):
            if tab[j] < tab[minimum]:
                minimum = j
        tab[i], tab[minimum] = tab[minimum], tab[i]
    return tab
```

```python
import random
import time


tab = []

for i in range(10000):
    tab.append(random.randint(1,100))
    

start = time.time()
SelectionSort(tab)
#SelectionSortB(tab)
stop = time.time()

print("Sortowanie trwało", stop-start)
```

### Insert Sort

```python
import random
tab = []

for i in range(10):
    tab.append(random.randint(1,25))
print(tab)
```

```python
def InsertSort(tab):
    for i in range(1, len(tab)): # przechodzę przez listę zaczynając od 2 elementu
        temp = tab[i] # Obecnie porównywany element w zmiennej "temp"
        j = i - 1
        
        # Przesuwamy elementy większe od "temp" o jedną pozycję w prawo,
        # aby zrobić miejsce na wstawienie elementu "temp"
        while j >= 0 and tab[j] > temp:
            tab[j+1] = tab[j]
            j -= 1
        tab[j+1] = temp    # Wstawia element "temp" na odpowiednie miejsce
        
    return tab  

InsertSort(tab)  
```

```python
def InsertSortKrok(tab, krok):
    for i in range(1, len(tab)):
        temp = tab[i]  # Zapisanie bieżącego elementu do wstawienia
        j = i - krok   # Zacznie porównywanie elementu oddalonego o "krok"
        
        # Przesuwa elementy mniejsze od temp o "krok"
        while j >= 0 and temp < tab[j]:  
            tab[j + krok] = tab[j]
            j = j - krok

        # Wstawia element temp na właściwe miejsce oddalone o "krok"
        tab[j+krok] = temp 
        
    return tab
```

```python
def InsertSortWar(tab):
    # Znajdowanie indeksu najmniejszego elementu w liście
    min_index = tab.index(min(tab))
    # Umieszczanie najmniejszego elementu na początku listy
    tab[0], tab[min_index] = tab[min_index], tab[0]
    
    for i in range(1, len(tab)):
        temp = tab[i]
        j = i - 1
        
        # Przesuwanie elementów większych od temp o jedno miejsce w prawo
        while tab[j] > temp:
            tab[j + 1] = tab[j]
            j -= 1
            
        # Wstawianie temp na odpowiednią pozycję
        tab[j + 1] = temp
    return tab
```
