### Quick Sort

```python
import random

tab = []

for _ in range(10):
    tab.append(random.randint(1,20))
print(tab)
    
def Partition(tab, start, stop):    # Partycjonowanie
    # Wybór losowego elementu jako pivot i zamiana go z ostatnim elementem
    los = random.randrange(start,stop)
    tab[los], tab[stop] = tab[stop], tab[los]
    p = tab[stop]  # Pivot staje się ostatnim elementem
    
    j = start # Pierwszy element podlisty
    
    # Iteracja przez listę w celu partcjonowania względem "pivot"
    for i in range(start, stop):
            # Warunek, który decyduje o przesunięciu elementu do lewej lub prawej strony pivot
            if (tab[i] <= p) or (tab[i] == p and i%2 == 1) :
                tab[i] , tab[j] = tab[j] , tab[i]
                j += 1
                
    # Umieszczenie pivot na jego końcowej pozycji
    tab[j], tab[stop] = tab[stop], tab[j]
    return j

Partition(tab,0,9)
```

```python
def QuickSort(tab, start, stop):
    if start >= stop:
        return
    pvt = Partition(tab, start, stop)
    QuickSort(tab, start, pvt-1)
    QuickSort(tab, pvt+1, stop)
    

QuickSort(tab, 0, 9)

#print(tab)
```

```python
def InsertSort(tab):
    for i in range(1, len(tab)):
        temp = tab[i]
        j = i - 1
        while j >= 0 and temp < tab[j]:
            tab[j+1] = tab[j]
            j -= 1
        tab[j+1] = temp
    return tab 
```
