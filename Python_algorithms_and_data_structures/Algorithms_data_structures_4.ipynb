### Merge Sort

```python
tab = [3,5,7,8,9,11,14,16,17,18,2,4,6,7,8,15,16,19]

def Merge(tab, start, mid, stop):
    t2 = [tab[k] for k in range(start,stop + 1)]
        
    i = 0
    j = mid - start
    
    for k in range(start, stop+1):
        if i == mid-start:
            tab[k] = t2[j]
            j += 1
        # Jeśli
        elif j > stop:
            tab[k] = t2[i]
            i += 1
            
        elif t2[i] <= t2[j]:
            tab[k] = t2[i]
            i += 1
            
        else:
            tab[k] = t2[j]
            j += 1
            
    return tab


Merge(tab,0,10,17)

```

```python
def merge_sort(tab):
    if len(tab) <= 1 :
        return tab
    
    middle = len(tab) // 2
    left_half = tab[:middle]
    right_half = tab[middle:]

    left_half = merge_sort(left_half)
    right_half = merge_sort(right_half)

    return merge(left_half, right_half)

def merge(left_half, right_half):
    i = j = 0
    result = []

    while i < len(left_half) and j < len(right_half):
        if left_half[i] < right_half[j]:
            result.append(left_half[i])
            i += 1
        else:   
            result.append(right_half[j])
            j += 1
    result += left_half[i:]
    result += right_half[j:]
    return result


lista = [1,5,123,33,12,553,1235,4578,33221,21,756]
merge_sort(lista)
```

```python
 # 1) Jzeli start >= stop -- nic nie rób
 # 2) Wylicz środek tablicy (mid)
 # 3) wywołaj rekurencyjnie merge sorta na lewej (start -> mid) i prawej (mid+1 - stop) 
 # 4) wywołaj merge - scal te dwie tablice
def MergeSort(tab, start, stop):
    if start >= stop:
        return
    mid = (start+stop) // 2
    MergeSort(tab,start, mid)
    MergeSort(tab,mid+1,stop)
    Merge(tab, start, mid+1, stop)

print(MergeSort(tab,0,17)) 
```
