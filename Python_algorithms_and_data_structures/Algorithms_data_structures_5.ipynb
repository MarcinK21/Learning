### FUNCKJA SILNIA

```python
def silnia1(n):   # SILNIA SEKWENCYJNIE
    wynik = 0 
    for i in range(1,n+1):
        wynik = wynik + i
    return wynik
```

```python
def silnia2(n):   # SILNIA REKURENCYJNIE
    if n == 0 :
        return 1
    return silnia2(n-1)*n
```

### CIĄG FIBONACCIEGO

```python
def fibS(n):    # FIBONACCI SEKWENCYJNIE
    a, b = 0, 1
    for i in range(1,n):
        a, b = b, a + b
    return b 
```

```python
fibS(50)
```

```python
def fibR(n):   # FIBONACCI REKURENCYJNIE
    if n == 0 :
        return 0
    elif n == 1 :
        return 1
    return fibbR(n-2)+fibbR(n-1)
```

### WIEŻA HANOI

```python
def Hanoi(n, start, cel):
    if n == 0 :
        return 0
    pom = 6 - start - cel 
    Hanoi(n-1, start, cel)
    print("z {} na {}".format(start,cel))
    Hanoi(n-1,pom, cel)
```

```python
Hanoi(5,1,3) 
```

```python
def FloodFill(x, y, nowy_kolor, stary_kolor = None):
    if nowy_kolor == stary_kolor:
        return

    if x < 0 or x > 9 or y < 0 or y > 9 :
        return

    if stary_kolor == None :
        stary_kolor = tab[x][y]

    if tab[x][y] != stary_kolor:
        return

    tab[x][y] = nowy_kolor

    FloodFill(x-1,y,nowy_kolor,stary_kolor)
    FloodFill(x+1,y,nowy_kolor,stary_kolor)
    FloodFill(x,y-1,nowy_kolor,stary_kolor)
    FloodFill(x,y+1,nowy_kolor,stary_kolor)
```

```python
# Inicjalizacja tablicy reprezentującej mapę pikseli
tab = []
for i in range(10):
    tab.append([0]*10)  # Wypełnienie tablicy zerami jako domyślnym kolorem

# Funkcja FloodFill do wypełniania obszaru
def FloodFill(x, y, nowy_kolor, stary_kolor=None):
    if nowy_kolor == stary_kolor:
        return
    if x < 0 or x > 9 or y < 0 or y > 9:
        return
    if stary_kolor is None:
        stary_kolor = tab[x][y]
    if tab[x][y] != stary_kolor:
        return

    tab[x][y] = nowy_kolor

    FloodFill(x-1, y, nowy_kolor, stary_kolor)
    FloodFill(x+1, y, nowy_kolor, stary_kolor)
    FloodFill(x, y-1, nowy_kolor, stary_kolor)
    FloodFill(x, y+1, nowy_kolor, stary_kolor)

# Przykład zastosowania funkcji FloodFill
# Ustawienie kolorów na mapie
tab[5][5] = 1
tab[5][6] = 1
tab[6][5] = 1
tab[6][6] = 1

# Wydruk mapy przed zastosowaniem FloodFill
print("Przed FloodFill:")
for row in tab:
    print(' '.join(map(str, row)))

# Użycie FloodFill do zmiany koloru obszaru
FloodFill(5, 5, 2)  # Zmiana koloru z 1 na 2

# Wydruk mapy po zastosowaniu FloodFill
print("\nPo FloodFill:")
for row in tab:
    print(' '.join(map(str, row)))
```
