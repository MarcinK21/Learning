### Drzewa BST

```python
import random
import time

# Klasa reprezentująca węzeł drzewa
class node(object):
    def __init__(self, number = None):
        self.number = number # Przechwouje wartość węzła
        self.left = None # Lewe dziecko węzła
        self.right = None # Prawe dziecko węzła


# Klasa reprezentująca drzewo binarne wyszukiwań
class tree(object):
    def __init__(self):
        self.dummy = node()  # Tworzy węzeł początkowy
    
    
    # Funkcja, metoda dodająca nowy węzeł do drzewa
    def Add(self, num, nd = None):
        if(nd == None):
            nd = self.dummy.right
        if(nd == None):    # jeśli drzewo było puste, dodaje pierwszy węzeł
            self.dummy.right  = node(num)
            return
        # Dodaje nowy węzeł 
        if(nd.number>num):
            if(nd.left == None):
                nd.left = node(num)
                return
            else:
                self.Add( num, nd.left)
                return
        else:
            if(nd.right == None):
                nd.right = node(num)
                return
            else:
                self.Add( num, nd.right)
                return
            
  
    # Drukuje elementy drzewa w kolejności
    def PrintIO(self, nd = "u"):       # IO - In order
        if(nd == "u"):
            nd = self.dummy.right
        if (nd == None):
            return
        if nd.left != None:
            self.PrintIO(nd.left)
        print(nd.number, end = ",")
        if nd.right != None:
            self.PrintIO(nd.right)
            
 
 
    def Suma(self, nd = "u"):    # Funkcja zwraca sumę wszystkich węzłów w drzewie
        if (nd == 'u'):
            nd = self.dummy.right
        if (nd == None):
            return 0
        return nd.number + self.Suma(nd.left) + self.Suma(nd.right)



    def Ilosc(self, nd = "u"):   # Funckja zwraca liczbę wezłów w drzewie
        if (nd == 'u'):
            nd = self.dummy.right
        if (nd == None):
            return 0
        return 1 + self.Ilosc(nd.left) + self.Ilosc(nd.right)

    
        
    def Wysokosc(self, nd = "u"):   # Funckja zwraca(oblicza) wysokość drzewa
        if (nd == 'u'):
            nd = self.dummy.right
        if (nd == None):
            return 0
        if self.Wysokosc(nd.left) > self.Wysokosc(nd.right):
            return 1 + self.Wysokosc(nd.left)
        return 1 + self.Wysokosc(nd.right)

 
 
    def Wysokosc2(self, nd = "u"):    # Drugi sposób na wyliczenie wysokości drzewa
        if (nd == 'u'):
            nd = self.dummy.right
        if (nd == None):
            return 0
        L = self.Wysokosc2(nd.left)
        R = self.Wysokosc2(nd.right)
        if L > R :
            return L + 1
        return R + 1



    def Myk(self, nd = "u", depth = 0):   # Metoda graficznie przedstawiająca strukturę drzewa
        if (nd == 'u'):
            nd = self.dummy.right
        if (nd == None):
            return 
        self.Myk(nd.right, depth +1)
        for _ in range(depth):
            print("  ", end=" ")
        print(nd.number)
        self.Myk(nd.left, depth + 1)



    def Najmniejszy(self, nd = "u"):         # Funkcja zwraca najmniejszy węzeł
        if (nd == "u"):
            nd = self.dummy.right
        if (nd == None):
            return None 
        if (nd.left == None):
            return nd
        return self.Najmniejszy(nd.left)   



t = tree()

how_much = 10



start_time = time.time()


for _ in range(0, how_much):
    t.Add(random.randint(1,20))

t.PrintIO()

print(" ")

elapsed_time = time.time()- start_time
#print(elapsed_time)


print("Suma wartości wszystkich węzłów: ", t.Suma())
print("Liczba węzłów: ", t.Ilosc())
print("Wysokość drzewa: ", t.Wysokosc())
print("Wysokość drzewa: ", t.Wysokosc2())
print("Najmniejsza wartość: ", t.Najmniejszy().number)
print("\nStruktura drzewa:" )
t.Myk()


```
