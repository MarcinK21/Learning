### Długość najdłuższego rosnącego podciągu ciągu liczb o losowej wartości

```python
def RandomValueList(od,do, ilość):
    import random
    tab = [random.randrange(od, do) for _ in range(ilość)]
    return tab

#Chcemy wyznaczyć najdłuższy podciąg rosnący
#f(n)-długość ściśle rosnącego ciągu, kończącego się wyrazem tab[n]
def f(n,tab):
    maks=1
    for k in range(0,n):
        if tab[k]<tab[n]: 
            w=f(k,tab)+1
            if w>maks:
                maks=w
    return maks

#Zapisujemy w tablicy mem poszczególne wartości naszych maksów
def f2(n,tab,mem):
    if mem[n]>0:
        return mem[n]
    maks=1
    for k in range(0,n):
        if tab[k]<tab[n]:
            w=f2(k,tab,mem)+1
            if w>maks:
                maks=w
    mem[n]=maks
    return maks

#Uzupełniamy tablice mem od dołu
#DNRP- długość najdłuższego rosnącego podciągu
def DNRP(tab):
    mem=[0]*len(tab)
    poz=[-1]*len(tab) # -1 oznacza że poprzednika nie ma, pod wartościami kryją się pozycje poprzedników tablicy tab
    for i in range(0, len(tab)):
        mem[i]=1
        for k in range(0,i):
            if tab[k]<tab[i]:
                if mem[k]+1>mem[i]:
                    mem[i]=mem[k]+1
                    poz[i]=k
    max_idx=0
    for i in range(0,len(tab)):
        if mem[i]> mem[max_idx]:
            max_idx=i
    z=max_idx
    wybrany=[]
    while z>-1:
        wybrany.append(z)
        z=poz[z]
    wybrany.reverse()
    for i in range(len(wybrany)):
        print(tab[wybrany[i]], end=",")
    print("\n")
    mem_maks=max(mem)
    return mem_maks

    
tab=RandomValueList(10,100,30)
mem=[0]*30
print(tab)


print("\n","pierwsza funckja f daje wynik:", f(29, tab), "\n")

print("\n","druga funckja f daje wynik:", f2(29, tab, mem), "\n")


#n=int(input("Podaj n: "))
#print("Długość podciągu rosnącego od n: ", f2(n,tab,mem))
print("Długość najdłuższego rosnącego podciągu: ", DNRP(tab))
```
