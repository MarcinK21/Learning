```python
import random
import time


def InsertSortKrok(tab, krok):
    for i in range(1, len(tab)):
        temp = tab[i]
        j = i - krok
        while j >= 0 and temp < tab[j]:
            tab[j + krok] = tab[j]
            j = j - krok

        tab[j+krok] = temp
        
    return tab


rozm = 100
tab=[random.randrange(1,99) for _ in range(rozm)]

def BucketSort(tab):
    # szykujemy kubełki
    Buckets = [[] for _ in range(10)]
    indeks = 0
    # jedziemy wzdłuż tab i wyznaczamy liczby do odpowiedniego kubełka
    for i in range(len(tab)) :
        numer = tab[i] // 10
        Buckets[numer].append(tab[i])
    # teraz jedziemy wzdłuż tablicy kubełków, sortujemy każdy i przepisujemy do tablicy tab
    #for lista in Buckets:
    #    InsertSortKrok(lista,1)
    for j in range(len(Buckets)):
        InsertSortKrok(Buckets[j],1)
        for k in range(len(Buckets[j])):
            tab[indeks] = Buckets[j][k]
            indeks = indeks + 1
    return tab
    


def BucketSortILE(tab, ile_kub):
    Buckets = [[] for _ in range(ile_kub)]
    indeks = 0
    

    for i in range(len(tab)) :
        numer = int(tab[i] // ( 100 / ile_kub))
        Buckets[numer].append(tab[i])
        
    for j in range(len(Buckets)):
        InsertSortKrok(Buckets[j],1)
        for k in range(len(Buckets[j])):
            tab[indeks] = Buckets[j][k]
            indeks = indeks + 1

    return Buckets



print(tab)
print(BucketSort(tab))
print(BucketSortILE(tab, 5))   
```
