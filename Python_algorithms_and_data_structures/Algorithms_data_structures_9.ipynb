### Counting Sort

```python
import random

tab = [random.randint(0,9) for _ in range(10)]

def CountingSort(tab):
    # Sprawdzenie czy tablica nie jest pusta
    # Teoretycznie jest niepotrzebne, ponieważ wcześniej tworzę tab, któe wiem że nie jest puste
    if not tab: 
        return tab
    
    
    #zorganizować tablicę liczników
    licznik = [0 for _ in range(max(tab) + 1)]

    #zliczyć wystąpienia poszczególnych cyfr
    for i in range(len(tab)):   
        licznik[tab[i]] += 1
    # Alternatynie:
    # for i in tab:
    # licznik[i] += 1    

    # skumulować tablicę liczników -> znaczy to tyle że każda pozycja zawiera sumę 
    # wartości wszystkich poprzednich pozycji, co pozwala na określenie pozycji
    # końcowej każdego elementu
    for j in range(1, len(licznik)):
        licznik[j] = licznik[j] + licznik[j-1]
    #   licznik += licznik[j-1]       -> to samo
    
    
    #zorganizować pustą tablicę
    Wynik = [-1]*len(tab)  # -1 jest tylko wartością tymczasową

    #przenieść ze starej do nowej w odpowiednie miejsca 
    for k in range(len(tab)-1,-1,-1):
        licznik[tab[k]] -= 1
        Wynik[licznik[tab[k]]] = tab[k]
    
    return Wynik


sorted_tab = CountingSort(tab)
print(sorted_tab)
```

### Radix Sort

```python
import random
import math

tab = [random.randint(0,9) for _ in range(10)]

# Funkcja GetDigit zwraca cyfrę liczby na danej pozycji (digit), licząc od końca.
def GetDigit(number, digit):
    number = str(number)  # Konwersja liczby na string, aby łatwo uzyskać dostęp do cyfr
    if digit < len(number):  # Sprawdzenie, czy podana pozycja jest w zakresie długości liczby
        return int(number[-1 - digit])  # Zwrócenie cyfry na danej pozycji
    else:
        return 0  # Zwrócenie 0, jeśli pozycja jest poza zakresem

# Alternatywna wersja GetDigit, która oblicza cyfrę matematycznie bez konwersji na string
def GetDigit2(number, digit):
    return int(number / (10**digit) % 10)  # Dzielenie przez 10^digit i modulo 10 zwraca cyfrę na danej pozycji

def CountingSort(tab, poz):
    liczniki = [0] * (max(tab) + 1)  # Tablica liczników dla każdej cyfry 
    
    # Zliczanie wystąpień każdej cyfry na danej pozycji w liczbie
    for i in range(len(tab)):
        liczniki[GetDigit(tab[i], poz)] += 1
    
    # Zsumowanie liczników, aby określić końcową pozycję każdej cyfry w posortowanej tablicy
    for i in range(1, len(liczniki)):
        liczniki[i] += liczniki[i-1]
    
    # Tworzenie posortowanej tablicy
    posort = [0] * len(tab)
    
    # Umieszczanie elementów w posortowanej tablicy na podstawie skumulowanych liczników
    for j in range(len(tab)-1, -1, -1):
        liczniki[GetDigit(tab[j], poz)] -= 1
        posort[liczniki[GetDigit(tab[j], poz)]] = tab[j]
    
    # Przepisanie posortowanej tablicy do oryginalnej tablicy
    for i in range(len(tab)):
        tab[i] = posort[i]




def RadixSort(tab):
    m = tab[0]  # Ustalamy maksymalną wartość tablicy na pierwszy element listy "tab"
    
    # Szukamy maksymalnej wartości w tablicy, aby wiedzieć, ile cyfr ma największa liczba
    for i in range(1, len(tab)):
        if tab[i] > m :
            m = tab[i]
            
    # Obliczamy liczbę cyfr maksymalnej wartości
    c = int(math.log(m, 10)) + 1
    
     # Wykonujemy sortowanie przez zliczanie dla każdej cyfry, zaczynając od najmniej znaczącej
    for dig in range(c):
        CountingSort(tab, dig)


RadixSort(tab)

print(tab)

# Znaleźć największą liczbę w tablicy
# Znaleźć jej liczbę 
```

```python
def radix_sort(nums):
    RADIX = 10  # Podstawa systemu liczbowego (dla liczb dziesiętnych to 10)
    placement = 1  # Określa, która cyfra liczby jest aktualnie sortowana (1 = jedności, 10 = dziesiątki, itd.)
    max_digit = max(nums)  # Znajdź maksymalną wartość w tablicy, aby wiedzieć, kiedy zakończyć sortowanie

    # Pętla wykonuje się, dopóki nie zostaną posortowane wszystkie cyfry największej liczby
    while placement < max_digit:
      buckets = [list() for _ in range(RADIX)]  # Tworzenie 10 kubełków dla cyfr od 0 do 9
      
      # Rozdzielenie liczb do odpowiednich kubełków na podstawie aktualnie sortowanej cyfry
      for i in nums:
        tmp = int((i / placement) % RADIX)
        buckets[tmp].append(i)
        
      a = 0  # Indeks w tablicy nums do ponownego umieszczania elementów z kubełków
      
      # Przejście przez wszystkie kubełki i umieszczenie ich elementów z powrotem w tablicy nums w posortowanej kolejności
      for b in range(RADIX):
        buck = buckets[b]
        
        for i in buck:
          nums[a] = i
          a += 1
          
      placement *= RADIX  # Przejście do kolejnej cyfry (np. z jedności do dziesiątek)
      
    return nums  # Zwrócenie posortowanej tablicy
  
user_input = input("Input numbers separated by a comma:\n").strip()
nums = [int(item) for item in user_input.split(',')]  # Konwersja danych wejściowych na listę liczb całkowitych
print(radix_sort(nums))  # Wywołanie funkcji sortującej i wyświetlenie wyników
```
