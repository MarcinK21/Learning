# Oczyszczanie danych - fillna()

```python
import pandas as pd
import numpy as np

fuel = pd.read_csv("./data/fuel.csv", low_memory = False,
                   usecols = ["Vehicle ID", "Year", "Make", "Model",
                              "Class", "Fuel Type", "Combined MPG (FT1)"],
                  index_col = "Vehicle ID")
fuel.head()
```

```python
fuel.loc[27705, 'Class'] = np.NaN
fuel.loc[26561, 'Class'] = np.NaN
fuel.loc[27550, 'Fuel Type'] = np.NaN
fuel.loc[27705, 'Combined MPG (FT1)'] = np.NaN
fuel.loc[27681, 'Combined MPG (FT1)'] = np.NaN

fuel.head()
```

```python
fuel.fillna(-1).head()
```

```python
replaceRules = {
    'Class':'---',
    'Fuel Type': '---',
    'Combined MPG (FT1)': -1
}
fuel.fillna(replaceRules).head()
```

```python
avgMPG = fuel['Combined MPG (FT1)'].mean()
avgMPG
```

```python
fuel['Class'].fillna('?', inplace= True)
fuel['Fuel Type'].fillna('?', inplace= True)
fuel['Combined MPG (FT1)'].fillna(avgMPG, inplace = True)
fuel.head()
```

```python
fuel['Combined MPG (FT1)'].fillna(method= 'ffill', inplace = True)
fuel.head()
```

```python
import pandas as pd
import numpy as np
fuel = pd.read_csv("./data/fuel.csv",
                   usecols=['Vehicle ID','Year','Make','Model','Class','Fuel Type','Combined MPG (FT1)','Start Stop Technology'],
                   index_col = 'Vehicle ID', low_memory=False)


fuel.head()
```

```python
listOfCars = ["Reanult", "Toyota", "Ford"]
isInList = fuel['Make'].isin(listOfCars)

isInList.head()
```

```python
fuel[isInList].head()
```

```python
isNoStartStopDefined = fuel["Start Stop Technology"].isnull()
isNoStartStopDefined.head()
```

```python
fuel[isNoStartStopDefined].head()
```

```python
isStartStopDefined = fuel["Start Stop Technology"].notnull()
isStartStopDefined.head()
```

```python
fuel[isStartStopDefined].head()
```

```python
mpg50_60 = fuel["Combined MPG (FT1)"].between(50,60)
mpg50_60.head()
```

```python
fuel[mpg50_60].head()
```

```python
print("There are", len(fuel[isInList]), "cars within the list", listOfCars)
print("There are", len(fuel[isNoStartStopDefined]), "cars where Start/Stop technology is not determined")
print("There are ", len(fuel[isStartStopDefined]), "cars where Start/Stop technology is determined")
print("There are", len(fuel[mpg50_60]), "cares with MPG value between 50 and 60")
```
