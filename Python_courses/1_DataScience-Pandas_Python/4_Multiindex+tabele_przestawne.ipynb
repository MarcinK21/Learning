```python
from datetime import datetime
from datetime import timedelta
import time
```

```python
df = pd.read_csv('./data/marathon_results_2016.csv', index_col='Bib',
usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'])
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df.head(1000)
```

```python
# Program sprawdzający każdy wiersz po kolei czy dane zgadzają się 
def is_valid_timedelta(value):
    try:
        pd.to_timedelta(value)
        return True
    except (ValueError, TypeError):
        return False

# Sprawdzenie kolumny '40K'
invalid_40K = df[~df['40K'].apply(is_valid_timedelta)]
print(invalid_40K)

# Sprawdzenie kolumny 'Half'
invalid_Half = df[~df['Half'].apply(is_valid_timedelta)]
print(invalid_Half)

''' Funkcja is_valid_timedelta, która próbuje przekonwertować wartość na timedelta i zwraca True, 
jeśli się uda. Funkcję przechodzi przez każdey element w kolumnach "40K" i "Half" i wyszukuje wiersze, 
dla których wartość jest nieprawidłowa. '''
```

```python
'''Wygląda na to że w niektórych kolumnach są wartości znaku "-", które oznacza brak danych
Zmienie te wartości na wartość NaN, a następnie pozbędę się tych wierszy'''

df['Half'] = df['Half'].replace('-', np.nan)
df['40K'] = df['40K'].replace('-', np.nan)

df = df.dropna(subset = ['Half', '40K'])
```

```python
# Teraz powinno wszystko zadziałać 
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df.head(1000)
```

```python
df.pivot_table(index = "Age", columns = "M/F", values = "TotalSeconds", aggfunc= "mean").head()
```

```python
df.pivot_table(index = "Age", columns = "M/F", values = ["HalfSeconds", "TotalSeconds"], aggfunc= "mean").head()
```

```python
df.pivot_table(index = "Age", columns = "M/F", values = ["HalfSeconds", "TotalSeconds"],
                aggfunc= "mean").swaplevel(axis = 'columns').head()
```

```python
df.pivot_table(index = "Age", columns = "M/F", values = ["HalfSeconds", "TotalSeconds"],
                aggfunc= "mean").swaplevel(axis = 'columns').sort_index(axis = 1).head()
```

```python
df.pivot_table(index = "Age", columns = "M/F", values = "HalfSeconds" , aggfunc= "count").head()
# W values może być dowolna rzecz, aggfunc i tak zlicza tylko ilość
# Ta tabela przestawna zlicza nam ilość zawodników 
```

```python
df.pivot_table(index = "Age", values = "HalfSeconds" , aggfunc= "count").head()
```

```python
df.pivot_table(index = "Country", values = "City", aggfunc= "count").head()
```

```python
df.pivot_table(index = "Country", values = "City", aggfunc= "count").sort_values('City',ascending= False).head()
```

```python
df.pivot_table(index = ["Country", "State", "City"], values = "TotalSeconds", aggfunc= "mean").head()
```

```python
df.pivot_table(index = ["Country", "State", "City"], values = "TotalSeconds",
                aggfunc= "mean").sort_values('TotalSeconds', ascending= False).head()
```
