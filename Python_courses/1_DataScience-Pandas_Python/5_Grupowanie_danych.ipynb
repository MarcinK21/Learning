# Metoda agg()

```python
products = pd.read_csv("./data/WA_Sales_Products_2012-14.csv")
products.head()
```

```python
groups = products.groupby(by = ["Retailer country", "Year"])
```

```python
groups["Revenue"].sum().head()
```

```python
groups["Quantity"].sum().head()
```

```python
groups["Gross margin"].mean().head()
```

```python
groups.Revenue.sum().head()
```

```python
groups.Quantity.sum().head()
```

```python
groups.agg({"Revenue" : "sum"}).head()
```

```python
groups.agg({"Gross margin" : "mean"}).head()
```

```python
groups.agg({"Revenue" : "sum",
            "Quantity" : "sum",
            "Gross margin" : "mean"}).head()
```

```python
groups.agg({"Revenue" : ["sum", "min", "max"],
            "Quantity" : ["sum", "min", "max"],
            "Gross margin" : "mean"}).head()
```

```python
my_aggregation = ["sum", "min", "max"]
```

```python
groups.agg({"Revenue" : my_aggregation,
            "Quantity" : my_aggregation,
            "Gross margin" : "mean"}).head()
```

# Przetwarzanie danych z obiektu grupy

```python
products = pd.read_csv("./data/WA_Sales_Products_2012-14.csv")
products.head()
```

```python
groups = products.groupby(by = "Retailer country")
```

```python
for country in groups:
    print(country[0])
```

```python
country_names = []

for country in groups:
    country_names.append(country[0])
```

```python
country_names
```

```python
for country in groups:
    print(country[1])
```

```python
for country in groups:
    print(country[0], len(country[1]))
```

```python
for country, country_data in groups:
    print(country, len(country_data))
```

```python
for country, country_data in groups:
    print(country,
    country_data["Revenue"].max() - country_data["Revenue"].min())
```

```python
for country, country_data in groups:
    print(country,
    country_data["Revenue"].max(),
    ", id_max:",
    country_data["Revenue"].idxmax())
```

```python
for country, country_data in groups:
    print(country,
    country_data["Revenue"].max(),
    country_data["Revenue"].idxmax(),
    country_data.loc[country_data["Revenue"].idxmax()])
```

```python
the_biggest = pd.DataFrame()


for country, country_data in groups:
    the_biggest = pd.concat([the_biggest, country_data.loc[[country_data["Revenue"].idxmax()]]])
```

```python
the_biggest.head()
```

```python
products.nlargest(1, "Revenue")
```

```python
the_biggest = pd.DataFrame()

for country, country_data in groups:
    the_biggest = pd.concat([the_biggest, country_data.nlargest(1, "Revenue")])
```
