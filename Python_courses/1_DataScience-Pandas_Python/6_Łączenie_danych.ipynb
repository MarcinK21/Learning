# Metoda merge

```python
cat1 = pd.read_csv("./data/northwind-mongo-master/categories_del_1.csv", usecols = ['CategoryID', 'CategoryName'])
cat2 = pd.read_csv("./data/northwind-mongo-master/categories_del_2.csv", usecols = ['CategoryID', 'CategoryName'])
cat1
```

```python
cat2
```

```python
cat1.merge(cat2)
```

```python
cat1.merge(cat2, on = 'CategoryID')
```

```python
cat1.merge(cat2, on = 'CategoryID', suffixes = ['_1', '_2'])
```

```python
cat1.merge(cat2, on = 'CategoryID', suffixes = ['_1', '_2'], how = 'outer', indicator = True)
```

```python
cat_merged = cat1.merge(cat2, on = 'CategoryID', suffixes = ['_1', '_2'], how = 'outer', indicator = True)
```

```python
filter = cat_merged['_merge'] == 'left_only'
cat_merged[filter]
```

```python
filter = cat_merged['_merge'] == 'right_only'
cat_merged[filter]
```

```python
filter = cat_merged['_merge'] == 'both'
cat_merged[filter]
```

```python
cat1.merge(cat2, on = 'CategoryID', suffixes =['_1', '_2'], how = 'left', indicator = True)
```

```python
cat1.merge(cat2, on = 'CategoryID', suffixes =['_1', '_2'], how = 'right', indicator = True)
```
