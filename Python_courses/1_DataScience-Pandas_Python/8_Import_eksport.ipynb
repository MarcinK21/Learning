# Zadania

```python
mcdonalds = pd.read_csv("./data/Mcdonalds.csv", usecols= ['Category', 'Item', 'Calories', 'TotalFat'])
mcdonalds.head()
```

```python
# nie eksportuj indeksu
mcdonalds.to_csv('mc-1.csv', index = False)
```

```python
# nie eksportuj indeksu, wszystkie teksty umieść w cudzysłowiu
mcdonalds.to_csv('mc-2.csv', index = False, quotechar = '"')
```

```python
# nie eksportuj indeksu, wszystkie teksty w cudzysłowiu, znakiem rozdzielającym część całkowitą od dziesiętnej ma być przecinek
mcdonalds.to_csv('mc-3.csv', index = False, quotechar = '"', decimal = ',')
```

```python
# 5 wszystko co powyżej + separatorem ma być przecinek + wyeksportowane mają zostać tyllko kolumny Item i TotalFat
mcdonalds.to_csv('mc-5.csv', index = False, quotechar = '"', decimal = ',', sep = ';', columns = ['Item', 'TotalFat'])
```
