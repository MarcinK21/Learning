#11. Rozwiązywanie układów równań

```python
A = np.array([[2, 4],
              [1, -1]])
print('A:', '\n', A, '\n')
B = np.array([[10],
             [-1]])
print('B:', '\n', B)
```

```python
A_inv = np.linalg.inv(A)
X = np.dot(A_inv, B)
print(X, '\n')
print('wymiar:',X.shape)
```
