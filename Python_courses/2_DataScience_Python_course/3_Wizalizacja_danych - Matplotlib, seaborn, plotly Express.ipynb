# 1. Wykres liniowy

```python
plt.plot([0,1], [0,3])
```

```python
plt.plot([0,1], [0,3], label = 'line', color = 'green')
plt.xlabel('oś X')
plt.ylabel('oś Y')
plt.title('Wykres liniowy')
_ = plt.legend()
```

```python
plt.plot([0, 1, 2], [0, 3, 2], label = 'line', color = 'green')
plt.xlabel('oś X')
plt.ylabel('oś Y')
plt.title('Wykres liniowy')
_ = plt.legend()
```

```python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)
plt.title('Wykres funkcji sin i cos')
plt.plot(x, x_sin)
_ = plt.plot(x, x_cos)
```

# 2. Tworzenie wielu wykresów

```python
plt.subplot(121)
plt.plot([0,1,2], [3,2,5])
plt.subplot(122)
plt.plot([0,1,2], [3,1,4], 'r--')
```

```python
fig = plt.figure(figsize = (12,8))
plt.subplot(131)   # oznacza to że chcemy mieć wyświetlane wykresy w 1 wierszu 3 kolumnach i odnosimy się do 1 wykresu
# 131 -> 1 wiersz 3 kolumny 1 - pierwszy wykres
plt.plot([0,1,2], [3,2,5])
plt.subplot(132)
plt.plot([0,1,2], [3,1,4], 'r--')
plt.subplot(133)
plt.plot([0,1,2], [3,1,4], 'ro')

fig.show()
```

```python
fig = plt.figure(figsize = (6,4))

for idx, img in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']):
    plt.subplot(221 + idx)
    plt.imshow(mpimg.imread(img))
    plt.axis('Off')
```

```python
fig = plt.figure(figsize = (10, 6))

for idx, img in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']):
    plt.subplot(141 + idx)
    plt.imshow(mpimg.imread(img))
    plt.axis('Off')
```

# Plotly: wykres kołowy i pierścieniowy

```python
from seaborn import load_dataset
df = load_dataset('diamonds')
df.head()
```

```python
dff = df.cut.value_counts()
dff = dff.reset_index()
dff.columns = ['type', 'count']
dff
```

```python
fig = go.Figure(
    data=go.Pie(labels=dff['type'], values=dff['count'])
)

fig.show()
```

```python
fig = go.Figure(
    data=go.Pie(labels=dff['type'], values=dff['count']),
    layout=go.Layout(title = 'Rozkład zmiennej cut', width=800)
)

fig.show()
```

```python
fig = go.Figure(
    data=go.Pie(labels=dff['type'], values=dff['count'], hole = 0.3), #hole = 0.3 tworzy wykres pierścieniowy
    layout=go.Layout(title = 'Rozkład zmiennej cut', width=800)
)

fig.show()
```

# Plotly: diagram Sankey'a

```python
data = [go.Sankey(node=dict(label=['Nonchurn_2018', 'Churn_2018', 'Nonchurn_2019', 'Churn_2019']),
                  link=dict(source=[0,0,1,1], #indeks odpowiadający etykiecie (labels)
                            target=[2,3,2,3],
                            value=[65,12,18,5]))]

fig = go.Figure(data=data,
                layout = go.Layout(width=800, height=400))
fig.show()
```

```python
data = [go.Sankey(node=dict(label=['Nonchurn_2018', 'Churn_2018', 'Nonchurn_2019', 'Churn_2019',
                                   'Nonchurn_2020','Churn_2020']),
                  link=dict(source=[0,0,1,1,2,3], #indeks odpowiadający etykiecie (labels)
                            target=[2,3,2,3,4,5],
                            value=[65,10,5,20,70,30]))]

fig = go.Figure(data=data,
                layout = go.Layout(width=800, height=400))
fig.show()
```
